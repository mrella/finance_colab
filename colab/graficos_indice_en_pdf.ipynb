```python
import yfinance as yf
import mplfinance as mpf
import pandas as pd
from fpdf import FPDF
from datetime import datetime, timedelta
import ta
import os
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
sns.set_theme(style="darkgrid", palette="muted", font_scale=1.1)

# Lista del NASDAQ 100 (resumida para ejemplo rápido, puedes ampliar)
nasdaq_100 = [
    "AAPL", "MSFT", "NVDA", "AMZN", "GOOGL", "META", "TSLA", "AMD"
]

# Fechas
end_date = datetime.today()
start_date = end_date - timedelta(days=180)

# Carpeta para gráficos
if not os.path.exists("charts"):
    os.makedirs("charts")

# PDF
pdf = FPDF()
pdf.set_auto_page_break(auto=True, margin=15)

for symbol in nasdaq_100:
    try:
        print(f"Procesando {symbol}...")
        df = yf.download(symbol, start=start_date, end=end_date, interval="1d")
        if df.empty:
            continue

        df.dropna(inplace=True)

        # Preparar datos
        df["EMA20"] = ta.trend.ema_indicator(df["Close"], window=20)
        df["EMA20_std"] = df["Close"].rolling(window=20).std()
        df["Upper_KC"] = df["EMA20"] + 2 * df["EMA20_std"]
        df["Lower_KC"] = df["EMA20"] - 2 * df["EMA20_std"]

        # MACD
        macd = ta.trend.macd(df["Close"])
        df["MACD"] = macd.macd()
        df["MACD_signal"] = macd.macd_signal()

        # Crear subplots
        apds = [
            mpf.make_addplot(df["Upper_KC"], color='orange'),
            mpf.make_addplot(df["Lower_KC"], color='orange'),
            mpf.make_addplot(df["MACD"], panel=2, color='blue', ylabel='MACD'),
            mpf.make_addplot(df["MACD_signal"], panel=2, color='red'),
        ]

        # Guardar gráfico
        chart_path = f"charts/{symbol}.png"
        mpf.plot(df, type='candle', style='yahoo', volume=True,
                 addplot=apds, title=f"{symbol} - 6 Meses",
                 savefig=dict(fname=chart_path, dpi=150, bbox_inches='tight'))

        # Añadir al PDF
        pdf.add_page()
        pdf.set_font("Arial", "B", 14)
        pdf.cell(0, 10, f"{symbol} - Gráfico de Velas, Volumen, MACD, Keltner", ln=True, align="C")
        pdf.image(chart_path, x=10, y=30, w=190)

    except Exception as e:
        print(f"Error con {symbol}: {e}")

# Guardar PDF
pdf.output("nasdaq100_analisis_tecnico.pdf")
print("✅ PDF generado: nasdaq100_analisis_tecnico.pdf")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import mplfinance as mpf

def calcular_koncorde(data):
    close = data['Close']
    volume = data['Volume']

    # Calcular OBV
    obv = (np.sign(close.diff()) * volume).fillna(0).cumsum()

    # Suavizados
    obv_ema = obv.ewm(span=20).mean()
    manos_fuertes = obv_ema - obv_ema.rolling(window=5).mean()
    manos_debiles = volume - volume.ewm(span=21).mean()

    return manos_fuertes, manos_debiles

def plot_koncorde(ticker, start="2023-01-01", end="2025-01-01"):
    data = yf.download(ticker, start=start, end=end)

    if data.empty:
        print("No hay datos.")
        return



    # Asegurar que el índice sea tipo datetime (por seguridad)
    data.index = pd.to_datetime(data.index)

    # Calcular Koncorde
    mf, md = calcular_koncorde(data)

    # Añadir al gráfico
    apds = [
        mpf.make_addplot(mf, panel=1, color='blue', width=1.2, ylabel='Manos Fuertes'),
        mpf.make_addplot(md, panel=1, type='bar', color='orange', alpha=0.5, width=0.7, ylabel='Manos Débiles')
    ]

    mpf.plot(
        data,
        type='candle',
        style='yahoo',
        title=f"{ticker} - Indicador Koncorde (Simplificado)",
        ylabel='Precio',
        addplot=apds,
        panel_ratios=(2, 1),
        volume=False,
        figscale=1.2
    )

# Prueba:
plot_koncorde("SPY", start="2024-06-01", end="2025-06-01")

```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def plot_spy_and_vix_ratio(start="2024-01-01", end="2025-01-01"):
    # Descargar datos
    spy = yf.download("SPY", start=start, end=end)
    vix = yf.download("^VIX", start=start, end=end)
    vix3m = yf.download("^VIX3M", start=start, end=end)

    # Asegurar fechas comunes
    common_dates = spy.index.intersection(vix.index).intersection(vix3m.index)
    spy = spy.loc[common_dates]
    vix = vix.loc[common_dates]
    vix3m = vix3m.loc[common_dates]

    # Calcular la relación VIX / VIX3M

    vix_ratio = vix["Close"] / vix3m["Close"]


    # Crear figura con subplots
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True, gridspec_kw={"height_ratios": [2, 1]})

    # Gráfico del SPY
    ax1.plot(spy.index, spy["Close"], label="SPY", color="blue")
    ax1.set_title("SPY y relación VIX / VIX3M")
    ax1.set_ylabel("Precio SPY")
    ax1.grid(True)
    ax1.legend()

    # Gráfico del ratio VIX/VIX3M
    ax2.plot(vix_ratio.index, vix_ratio, label="VIX / VIX3M", color="purple")
    ax2.axhline(1, color="gray", linestyle="--", linewidth=1)
    ax2.set_ylabel("Relación VIX / VIX3M")
    ax2.set_xlabel("Fecha")
    ax2.grid(True)
    ax2.legend()

    plt.tight_layout()
    plt.show()

# Ejecutar
plot_spy_and_vix_ratio("2024-06-01", "2025-06-20")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def calculate_kama(prices, window=10, fast=2, slow=30):
    """
    Cálculo simplificado de KAMA.
    """
    change = abs(prices - prices.shift(window))
    volatility = prices.diff().abs().rolling(window).sum()
    er = change / volatility

    fast_sc = 2 / (fast + 1)
    slow_sc = 2 / (slow + 1)
    sc = (er * (fast_sc - slow_sc) + slow_sc) ** 2

    kama = pd.Series(index=prices.index, dtype='float64')
    kama.iloc[window] = prices.iloc[window]

    for i in range(window + 1, len(prices)):
        kama.iloc[i] = kama.iloc[i - 1] + sc.iloc[i] * (prices.iloc[i] - kama.iloc[i - 1])

    return kama

def plot_spy_with_kama(start="2024-01-01", end="2025-01-01"):
    # Descargar datos
    data = yf.download("SPY", start=start, end=end)
    close = data["Close"]

    # Calcular KAMA
    kama = calculate_kama(close, window=10, fast=2, slow=30)

    # Graficar
    plt.figure(figsize=(12, 6))
    plt.plot(close, label="SPY Close", color="blue")
    plt.plot(kama, label="KAMA", color="orange", linewidth=2)
    plt.title("SPY con KAMA (Kaufman's Adaptive Moving Average)")
    plt.xlabel("Fecha")
    plt.ylabel("Precio")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Ejecutar
plot_spy_with_kama("2024-06-01", "2025-06-20")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def calculate_kama(prices, window=10, fast=2, slow=30):
    change = abs(prices - prices.shift(window))
    volatility = prices.diff().abs().rolling(window).sum()
    er = change / volatility

    fast_sc = 2 / (fast + 1)
    slow_sc = 2 / (slow + 1)
    sc = (er * (fast_sc - slow_sc) + slow_sc) ** 2

    kama = pd.Series(index=prices.index, dtype='float64')
    kama.iloc[window] = prices.iloc[window]

    for i in range(window + 1, len(prices)):
        kama.iloc[i] = kama.iloc[i - 1] + sc.iloc[i] * (prices.iloc[i] - kama.iloc[i - 1])

    return kama

def plot_spy_with_kama_and_volume(start="2024-01-01", end="2025-01-01"):
    # Descargar datos
    data = yf.download("SPY", start=start, end=end)
    close = data["Close"]
    volume = data["Volume"]

    # Calcular KAMA
    kama = calculate_kama(close, window=10, fast=2, slow=30)

    # Crear figura con 2 subgráficos (precio + volumen)
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True,
                                   gridspec_kw={"height_ratios": [3, 1]})

    # Gráfico de precios con KAMA
    ax1.plot(close, label="SPY Close", color="blue")
    ax1.plot(kama, label="KAMA", color="orange", linewidth=2)
    ax1.set_title("SPY con KAMA (Kaufman's Adaptive Moving Average)")
    ax1.set_ylabel("Precio")
    ax1.legend()
    ax1.grid(True)

    # Gráfico de volumen
    ax2.bar(volume.index, volume, color="gray", width=1)
    ax2.set_ylabel("Volumen")
    ax2.set_xlabel("Fecha")
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

# Ejecutar
plot_spy_with_kama_and_volume("2024-06-01", "2025-06-20")
```
