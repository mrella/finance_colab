```python
import requests

def get_latest_pelosi_trades():
    # Endpoint para transacciones de Pelosi (P000197)
    url = "https://api.poliapi.com/v1/trades?politician=P000197"

    resp = requests.get(url)
    resp.raise_for_status()  # Lanza excepción si hay error

    trades = resp.json().get("data", [])  # Ajusta según formato real

    # Filtrar compras recientes
    recent_buys = [t for t in trades if t.get("type") == "buy"]

    # Mostrar los campos clave
    for t in recent_buys:
        date   = t.get("traded") or t.get("date")
        ticker = t.get("ticker") or t.get("traded_issuer_ticker")
        amount = t.get("size") or t.get("amount_range")
        print(date, ticker, amount)

if __name__ == "__main__":
    get_latest_pelosi_trades()
```
